# Les conditions

```python
""" Structure générale d'une condition if """
if mon_booleen:
   # exécuter le code quand mon_booleen est vrai
else:
   # exécuter le code quand mon_booleen est faux
```

Rappel: Les booléens peuvent être combinés avec des opérateurs logiques (and, or ..) ou issus de comparaisons (<, >, == ...)

```python
""" Exemple """

ensoleille = False

if ensoleille:
   print("on va à la plage !")
else:
   print("on reste à la maison !")
```

```python
""" Il est également possible d'ajouter d'autres conditions"""

ensoleille = False
neige = True

if ensoleille:
   print("on va à la plage !")
elif neige:
   print("on fait un bonhomme de neige")
else:
   print("on reste à la maison !")

a = 8

if a > 2:
   print(a)
if a < 7:
   print(a)
else:
   print("nothing")

```

# Les boucles

## La boucle for

```python
#Le principal interet de la boucle for est de parcourir des listes
liste_apprenants = ["Pascal", "Azel", "Tony", "Vladimir", "Sabrina"]

for element in liste_apprenants:
    print(element, " présent")
```

On peut également s'en servir avec la fonction range, cela permet de parcourir une liste d'entiers

```python
print(type(range(5)))
print(range(5))
```

```python
for i in range(10):
    print(i)
```

```python
# la fonction range peut prendre plusieurs arguments:
premier_element = 2
dernier_element_plus_1 = 11
pas = 2 # on compte tous les combien

for j in range(premier_element,dernier_element_plus_1,pas):
    print(f"j = {j}")

premier_element = 11
dernier_element_plus_1 = 7
pas = -3 # on compte tous les combien

for k in range(premier_element,dernier_element_plus_1,pas):
    print(f"k = {k}")   
```

## Boucle While

```python
# La structure générale de la boucle while est la suivante:

i = 0
condition = 10 > i

while condition:
    print(f"la condition est encore vraie car i vaut {i}")
    i += 1 # on appelle cela l'incrementation
    condition = 10 > i
```

```python
# l'incrémentation permet de ne pas créer de boucle infinie
i=11
condition = 12 > i
while condition:
    print("ceci est une boucle infinie")
    
```

# Bonus le mot With

On retrouve le mot with dans un usage particulier qui est le travail avec un fichier texte.
Quand on travaille avec un fichier texte, il faut l'ouvrir pour accéder à son contenu et ne surtout pas oublier de le fermer sinon cela peut entrainer un long enchainement de bugs.
Sauf que si vous faites une erreur dans votre code, le code s'arretra et ne lira pas la partie ou vous fermez votre fichier. Pour s'assurer de ne pas se retrouver dans ce cas, on ouvrira systématiquement nos fichiers avec with.

```python
with open("file.txt") as fichier:
   for ligne in fichier:
      # faire quelque chose avec une ligne
      print(ligne)

# car attention:
with open("file.txt") as fichier:
    print(fichier)
```

```python
# On peut aussi écrire dans un fichier:

with open('file.txt', 'w') as file: # W pour écire
    file.write('hello world !')
```
